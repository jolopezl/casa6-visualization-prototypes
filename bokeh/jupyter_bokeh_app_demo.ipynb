```python
def draw_xy(vis, exclude=[]):
    telescope, names, ids, xpos, ypos, stations = getPlotantsAntennaInfo(vis, False, exclude, False)
#     print(telescope)
#     print(names)
#     print(ids)
#     print(stations)
    source = bm.ColumnDataSource(data=dict(xp=xpos,
                                             yp=ypos,
                                             labels=names))

    # create a new plot with default tools, using figure
    p = bp.figure(title=telescope, plot_width=400, plot_height=400)
    p.scatter(xpos, ypos, size=10, line_color="red", fill_color="red", fill_alpha=0.5)
    
    if telescope == 'VLBA':
        labelx = 'Longitude (deg)'
        labely = 'Latitude (deg)'
    else:
        # use m or km units
        units = ' (m)'
        if np.median(xpos) > 1e6 or np.median(ypos) > 1e6:
            xpos /= 1e3
            ypos /= 1e3
            units = ' (km)'
        labelx = 'X' + units
        labely = 'Y' + units
    
    p.xaxis[0].axis_label = labelx
    p.yaxis[0].axis_label = labely
    
    labels = bm.LabelSet(x='xp', y='yp', text='labels', level='glyph',
                      x_offset=10, y_offset=10, source=source,
                      render_mode='canvas')
    
    p.add_layout(labels)
    # show the results
    bp.show(p)
```
